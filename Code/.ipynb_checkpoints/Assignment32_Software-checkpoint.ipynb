# Attributes cretaion with Amount, Frequency and Recency data

```python
data['Amount'] = data['Quantity']*data['UnitPrice']
rfm_m = data.groupby('CustomerID')['Amount'].sum()
rfm_m = rfm_m.reset_index()
rfm_m.head()
```

# New Attribute creation with Frequency

```python
rfm_f = data.groupby('CustomerID')['InvoiceNo'].count()
rfm_f = rfm_f.reset_index()
rfm_f.columns = ['CustomerID', 'Frequency']
rfm_f.head()
```

# Merging two differnt dataflows

```python
rfm = pd.merge(rfm_m, rfm_f, on='CustomerID', how='inner')
rfm.head()
```

# New Attribute creation with Recency

# Convert  datetime into appropiate datatype

```python
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'],format='%d-%m-%Y %H:%M')
```

```python
max_date = max(data['InvoiceDate'])
max_date
```

# Calculation of the difference of max date and transaction date

```python
data['Diff'] = max_date - data['InvoiceDate']
data.head()
```

# Computation of the last transaction date to achieve the recency of the customers

```python
rfm_p = data.groupby('CustomerID')['Diff'].min()
rfm_p = rfm_p.reset_index()
rfm_p.head()
```

# Extract only the number of days 

```python
rfm_p['Diff'] = rfm_p['Diff'].dt.days
rfm_p.head()
```

# combining the differnt dataframes to achieve the final RFM dataframe

```python
rfm = pd.merge(rfm, rfm_p, on='CustomerID', how='inner')
rfm.columns = ['CustomerID', 'Amount', 'Frequency', 'Recency']
rfm.head()
```

# Dealing with the Outliers to generate outliers variable distribution

```python
attributes = ['Amount','Frequency','Recency']
plt.rcParams['figure.figsize'] = [10,8]
sns.boxplot(data = rfm[attributes], orient="v", palette="Set2" ,whis=1.5,saturation=1, width=0.7)
plt.title("Outliers Variable Distribution", fontsize = 14, fontweight = 'bold')
plt.ylabel("Range", fontweight = 'bold')
plt.xlabel("Attributes", fontweight = 'bold')
```

```python

Q1 = rfm.Amount.quantile(0.05)
Q3 = rfm.Amount.quantile(0.95)
IQR = Q3 - Q1
rfm = rfm[(rfm.Amount >= Q1 - 1.5*IQR) & (rfm.Amount <= Q3 + 1.5*IQR)]


Q1 = rfm.Recency.quantile(0.05)
Q3 = rfm.Recency.quantile(0.95)
IQR = Q3 - Q1
rfm = rfm[(rfm.Recency >= Q1 - 1.5*IQR) & (rfm.Recency <= Q3 + 1.5*IQR)]


Q1 = rfm.Frequency.quantile(0.05)
Q3 = rfm.Frequency.quantile(0.95)
IQR = Q3 - Q1
rfm = rfm[(rfm.Frequency >= Q1 - 1.5*IQR) & (rfm.Frequency <= Q3 + 1.5*IQR)]
```

# Rescaling different attributes

```python
rfm_data = rfm[['Amount', 'Frequency', 'Recency']]


scaler = StandardScaler()


rfm_data_scaled = scaler.fit_transform(rfm_data)
rfm_data_scaled.shape
```

```python
rfm_data_scaled = pd.DataFrame(rfm_data_scaled)
rfm_data_scaled.columns = ['Amount', 'Frequency', 'Recency']
rfm_data_scaled.head()
```

```python
rfm_data_scaled = pd.DataFrame(rfm_data_scaled)
rfm_data_scaled.columns = ['Amount', 'Frequency', 'Recency']
rfm_data_scaled.head()
```

# k-means with some of the arbitraries

```python
kmeans = KMeans(n_clusters=4, max_iter=50)
kmeans.fit(rfm_data_scaled)
```

```python
kmeans.labels_
```

# Elbow-curve versus SSD to genrate clusters versus ssd diagram

```python
ssd = []
range_n_clusters = [2, 3, 4, 5, 6, 7, 8]
for num_clusters in range_n_clusters:
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50)
    kmeans.fit(rfm_data_scaled)
    
    ssd.append(kmeans.inertia_)
    

plt.plot(ssd, marker='o')
plt.title('Clusters Vs SSD')
plt.xlabel('No of Clusters')
plt.ylabel('Intertia')
```

```python

range_n_clusters = [2, 3, 4, 5, 6, 7, 8]

for num_clusters in range_n_clusters:
    

    kmeans = KMeans(n_clusters=num_clusters, max_iter=50)
    kmeans.fit(rfm_data_scaled)
    
    cluster_labels = kmeans.labels_
    

    silhouette_avg = silhouette_score(rfm_data_scaled, cluster_labels)
    print("For n_clusters={0}, the silhouette score is {1}".format(num_clusters, silhouette_avg))
```

```python
kmeans = KMeans(n_clusters=3, max_iter=50)
kmeans.fit(rfm_data_scaled)
```

```python
 kmeans.labels_
```

```python
rfm['Cluster_Id'] = kmeans.labels_
rfm.head()
```

# single likage

```python
mergings = linkage(rfm_data_scaled, method="single", metric='euclidean')
dendrogram(mergings)
plt.show()
```

# Complete linkage

```python
mergings = linkage(rfm_data_scaled, method="complete", metric='euclidean')
dendrogram(mergings)
plt.show()
```

# Average likage result

```python
mergings = linkage(rfm_data_scaled, method="average", metric='euclidean')
dendrogram(mergings)
plt.show()
```

```python
cluster_labels = cut_tree(mergings, n_clusters=3).reshape(-1, )
cluster_labels
```

```python
rfm['Cluster_Labels'] = cluster_labels
rfm.head()
```

```python
sns.boxplot(x='Cluster_Labels', y='Amount', data=rfm)
```
